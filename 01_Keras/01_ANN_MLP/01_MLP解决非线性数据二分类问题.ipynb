```python
import numpy as np
import pandas as pd
```

```python
random_state = 666
np.random.seed(random_state)
```

```python
# load data
data_frame = pd.read_csv('./data_01.csv')
```

```python
data_frame.columns.values.tolist()
```

```python
data = data_frame.values
X = data[:, :2]
y = data[:, 2]

# shuffle
shuffle_indexes = np.random.permutation(data.shape[0])
X = X[shuffle_indexes]
y = y[shuffle_indexes]
```

```python
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
# train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=random_state)
```

```python
from keras.models import Sequential
from keras.layers import Dense

# 建立MLP模型
mlp_model = Sequential()
mlp_model.add(layer=Dense(units=20, input_dim=2, activation='sigmoid'))
mlp_model.add(layer=Dense(units=1, activation='sigmoid'))

# 配置模型损失函数和最优化方法
mlp_model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
# 查看模型结构
mlp_model.summary()
```

```python
# 模型训练
mlp_model.fit(X_train, y_train, epochs=3000, verbose=1)
```

```python
from sklearn.metrics import accuracy_score

# 训练集
y_train_predict = mlp_model.predict_classes(X_train)
train_score = accuracy_score(y_true=y_train, y_pred=y_train_predict)
train_score
```

```python
# 测试集
y_test_predict = mlp_model.predict_classes(X_test)
test_score = accuracy_score(y_true=y_test, y_pred=y_test_predict)
test_score
```

```python
import matplotlib.pyplot as plt

X_plot = X
y_plot = mlp_model.predict_classes(X).reshape(X_plot.shape[0], )
plt.scatter(X_plot[y_plot == 0, 0], X_plot[y_plot == 0, 1], color='r')
plt.scatter(X_plot[y_plot == 1, 0], X_plot[y_plot == 1, 1], color='g')
plt.show()
```
